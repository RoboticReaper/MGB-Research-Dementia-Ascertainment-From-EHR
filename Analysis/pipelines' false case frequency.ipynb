```python
import pandas as pd
import os
```

```python
false_aggregate = pd.DataFrame()
data_path = "data"
files = os.listdir(data_path)

for file in files:
    full_path = os.path.join(data_path, file)
    original_df = pd.read_csv(full_path)
    original_df['pipeline'] = file[:-4]
    false_aggregate = pd.concat([false_aggregate, original_df], ignore_index=True)

print(false_aggregate)
```

```python
result = pd.DataFrame()

for empi in false_aggregate['empi'].unique():
    this_patient = false_aggregate[false_aggregate['empi'] == empi]
    new_row = {'empi': empi, 'label': this_patient.iloc[0]['label'], 'pipelines_wrong': 0}

    for index, record in this_patient.iterrows():
        new_row[f'pipeline_{record["pipeline"]}_prediction'] = record['prediction']
        if record['prediction'] != record['label']:
            new_row['pipelines_wrong'] += 1

    result = pd.concat([result, pd.DataFrame([new_row])], ignore_index=True)

result_desc = result.sort_values(by='pipelines_wrong', ascending=False)
result_desc.to_csv("false.csv", index=False)
print(result_desc)
```
