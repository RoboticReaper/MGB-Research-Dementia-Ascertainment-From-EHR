```python
import pandas as pd
import os
```

```python
df = pd.read_csv("rule_based_classification_results.csv")
```

```python
r1_tp = 0
r1_fp = 0
r1_fn = 0
r1_tn = 0

r2_tp = 0
r2_fp = 0
r2_fn = 0
r2_tn = 0

for _, row in df.iterrows():
    label = row['human_label']
    r1 = row['rule1_label']
    r2 = row['rule2_label']

    if label == 1:
        if r1 == 1:
            r1_tp += 1
        else:
            r1_fn += 1
    else:
        if r1 == 1:
            r1_fp += 1
        else:
            r1_tn += 1

    if label == 1:
        if r2 == 1:
            r2_tp += 1
        else:
            r2_fn += 1
    else:
        if r2 == 1:
            r2_fp += 1
        else:
            r2_tn += 1

print(f"r1 tp: {r1_tp} fp: {r1_fp} tn: {r1_tn} fn: {r1_fn}")
print(f"r2 tp: {r2_tp} fp: {r2_fp} tn: {r2_tn} fn: {r2_fn}")
```
